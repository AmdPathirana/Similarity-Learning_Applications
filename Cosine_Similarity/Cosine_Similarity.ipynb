 Python code snippet to calculate the cosine similarity between two vectors using numpy

```python
import numpy as np

def cosine_similarity(vector1, vector2):
    dot_product = np.dot(vector1, vector2)
    norm_vector1 = np.linalg.norm(vector1)
    norm_vector2 = np.linalg.norm(vector2)
    return dot_product / (norm_vector1 * norm_vector2)

# Example usage:
vector1 = np.array([1, 2, 3])
vector2 = np.array([4, 5, 6])

similarity = cosine_similarity(vector1, vector2)
print("Cosine Similarity:", similarity)
```

<font color='green'> 
Cosine similarity between vector1 and vector2 using numpy's dot function for the dot product and linalg.norm for the vector norms.

## Feature Extraction [ VGG16 ]

```python
from skimage import io
from skimage.transform import resize
from tensorflow.keras.applications.vgg16 import VGG16, preprocess_input
from tensorflow.keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt
from skimage import io


def load_and_preprocess_image(image_path, target_size):
    img = io.imread(image_path)
    img = resize(img, target_size)
    img = img.astype(np.float32)
    img = preprocess_input(img)
    return img

def extract_features(image_path):
    img = load_and_preprocess_image(image_path, target_size=(224, 224))
    model = VGG16(weights='imagenet', include_top=False)
    img = np.expand_dims(img, axis=0)
    features = model.predict(img)
    features = features.flatten()
    return features

def cosine_similarity(vector1, vector2):
    dot_product = np.dot(vector1, vector2)
    norm_vector1 = np.linalg.norm(vector1)
    norm_vector2 = np.linalg.norm(vector2)
    return dot_product / (norm_vector1 * norm_vector2)



image_path1 = 'img/1.jpg'
image_path2 = 'img/2.webp'


features1 = extract_features(image_path1)
features2 = extract_features(image_path2)

similarity = cosine_similarity(features1, features2)
print("Cosine Similarity:", similarity)

# Load images
image1 = io.imread('img/1.jpg')
image2 = io.imread('img/2.webp')

# Display image1
plt.subplot(1, 2, 1)
plt.imshow(image1)
plt.title('Image 1')

# Display image2
plt.subplot(1, 2, 2)
plt.imshow(image2)
plt.title('Image 2')

plt.show()
```

## Feature Extraction [ ResNet50 ]

```python
import matplotlib.pyplot as plt
from skimage import io
import numpy as np
from tensorflow.keras.applications.resnet50 import ResNet50, preprocess_input
from tensorflow.keras.preprocessing import image
from tensorflow.keras.models import Model
import matplotlib.pyplot as plt
from skimage import io

def load_and_preprocess_image(image_path, target_size):
    img = io.imread(image_path)
    img = resize(img, target_size)
    img = img.astype(np.float32)
    img = preprocess_input(img)
    return img

def extract_features(image_path):
    img = load_and_preprocess_image(image_path, target_size=(224, 224))
    base_model = ResNet50(weights='imagenet', include_top=False, pooling='avg')
    model = Model(inputs=base_model.input, outputs=base_model.output)
    img = np.expand_dims(img, axis=0)
    features = model.predict(img)
    features = features.flatten()
    return features

def cosine_similarity(vector1, vector2):
    dot_product = np.dot(vector1, vector2)
    norm_vector1 = np.linalg.norm(vector1)
    norm_vector2 = np.linalg.norm(vector2)
    return dot_product / (norm_vector1 * norm_vector2)


image_path1 = 'img/4.jpeg'
image_path2 = 'img/5.jpg'

features1 = extract_features(image_path1)
features2 = extract_features(image_path2)

similarity = cosine_similarity(features1, features2)
print("Cosine Similarity:", similarity)


# Load images
image1 = io.imread(image_path1)
image2 = io.imread(image_path2)

# Display image1
plt.subplot(1, 2, 1)
plt.imshow(image1)
plt.title('Image 1')

# Display image2
plt.subplot(1, 2, 2)
plt.imshow(image2)
plt.title('Image 2')

plt.show()
```
